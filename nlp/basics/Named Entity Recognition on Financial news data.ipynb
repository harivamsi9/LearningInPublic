```python
def load_article_file(file_path):
    try:
        with open(file_path, "r", encoding="utf-8") as file:
            text = file.read()
        return text
    
    except FileNotFoundError:
        print(f"File not found: {file_path}")
    except Exception as e:
        print(f"An error occurred while loading the file: {e}")
        return None
    
    
article_file_path = "/Users/harivamsi/Desktop/Projects/hashnode/nlp/newsarticle.txt"  # Replace this with the actual file path
text = load_article_file(article_file_path)
print(text)
```

```python
import spacy
nlp = spacy.load('en_core_web_sm')
doc = nlp(text)
```

```python
# Tokenization
for token in doc:
    print(token.text)
```

```python
# Tokenize by sentenses
for sent in doc.sents:
    print(sent)
```

```python
# Lemmatization
for token in doc:
    print(token.text, " | ",  token.lemma_)
```

```python
# POS Tagging
for token in doc:
    print(token.text, " | ", token.lemma_, " | ", token.pos_)
```

```python
# NER
for ent in doc.ents:
    print(ent.text, " | ", ent.label_)
```

```python
from spacy import displacy
displacy.render(doc, style='ent')
```

```python
#Create the EntityRuler
ruler = nlp.add_pipe("entity_ruler", before="ner")
#List of Entities and Patterns
patterns = [
                {"label": "STOCK", "pattern": "^GSPC"},
                {"label": "STOCK", "pattern": "^DJI"},
                {"label": "STOCK", "pattern": "^IXIC"},
                {"label": "STOCK", "pattern": "CAT"},
                {"label": "STOCK", "pattern": "PFE"},
                {"label": "STOCK", "pattern": "UBER"},
                {"label": "STOCK", "pattern": "AAPL"},
                {"label": "STOCK", "pattern": "AMZN"},
                {"label": "ORG", "pattern": "Pfizer"},
                {"label": "ORG", "pattern": "Uber"},
    
            ]
ruler.add_patterns(patterns)
doc = nlp(text)
# NER
for ent in doc.ents:
    print(ent.text, " | ", ent.label_)
    
from spacy import displacy
displacy.render(doc, style='ent')
```
